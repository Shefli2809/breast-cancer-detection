## Importing Libraries and Datasets

```
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# importing dataset
dataset = pd.read_csv('data.csv')
```

```
dataset.mean()
```

```
# replacing NaN values with the mean values
dataset =  dataset.fillna(dataset.mean())
```

```
# plotting heatmap to visualize the number of NaN's in the dataset

plt.figure(figsize=(20,20))
sns.heatmap(dataset.isnull(), yticklabels = False)
plt.show()
```

## Preparing data before training

```
# separating features and dependent variables
X = dataset.iloc[: , 1:-1].values
y = dataset.iloc[: , -1].values
y = y.reshape(-1,1)
```

```
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```
# encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y) # 0 = benign, 1 = malignant
```

```
# splitting into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3) # 80/20
```

## Training and Testing the model

```
import xgboost as xgb
classifier = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 50, n_estimators = 100)
classifier.fit(X_train, y_train)
```

```
# predicting test data
y_pred = classifier.predict(X_test)
```

```
# plotting confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
plt.show()
```

```
# Accuracy scores
print(f"Training set score: {classifier.score(X_train, y_train)*100:.2f}")
print(f"Test set score: {classifier.score(X_test, y_test)*100:.2f}")
```
